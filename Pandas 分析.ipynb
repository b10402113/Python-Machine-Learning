```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
#108年公司設立登記清冊
df = pd.read_csv("http://data.gcis.nat.gov.tw/od/file?oid=3D064F20-FDC8-44DF-8154-395D19FC8891")
```

```python
from matplotlib.font_manager import FontProperties

import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei'] 
plt.rcParams['axes.unicode_minus'] = False
```

```python
df.head()
```

```python
#新增地區欄位
df['地區'] = df['公司所在地'].str[:3]
```

```python
#將企業的資本額按造資本額去分企業分佈圖
df2 = df[:100].sort_values(by = ['資本額'], ascending=False)
```

```python
#108年前二十大新公司的資本額排序
df2[:20].plot.bar(x='公司名稱', y=['資本額'])
```

```python
locations = {}
```

```python
#製作地點字典
for i in range(0,len(df)):
    if df['地區'][i] in locations:
        locations[df['地區'][i]] += 1
    else:
        locations[df['地區'][i]] = 0
locations
```

```python
#顯示台灣2019年公司創業的分布圖
labels = locations.keys()
size = locations.values()
plt.pie(size , labels = labels,autopct='%1.1f%%');
```
