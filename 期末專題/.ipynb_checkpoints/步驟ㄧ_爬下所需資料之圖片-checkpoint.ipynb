<font color=#1a2933  size=5 face="黑体">一、引入必須套件</font>


```python
import flickrapi
import urllib.request
from PIL import Image
```



<font color=#1a2933  size=5 face="黑体">二、Flickr-API</font>

<font color=#1a2933  size=3 face="黑体">
這邊使用FlickrAPI去下載flickr的照片
</font>

```python
flickr=flickrapi.FlickrAPI('c6a2c45591d4973ff525042472446ca2', '202ffe6f387ce29b', cache=True)
```



<font color=#1a2933  size=3 face="黑体">
    
flickr.walk會去將flickr的圖片連結存入photos中，只要給適當的keyword
</font>

```python
keyword = 'statue of liberty'

photos = flickr.walk(text=keyword,
                     tag_mode='all',
                     tags=keyword,
                     extras='url_c',
                     per_page=1000,           
                     sort='relevance')
```

```python
urls = []
for i, photo in enumerate(photos):
    print (i)
    url = photo.get('url_c')
    urls.append(url)
    
    # get 50 urls
    if i > 1000:
        break
```

```python
print (urls)

# Download image from the url and save it to '00001.jpg'
for i in range(0,900):
    if(urls[i] != None):
        urllib.request.urlretrieve(urls[i], '0000'+str(i)+'.jpg')
        # Resize the image and overwrite it
        image = Image.open('0000'+str(i)+'.jpg') 
        image = image.resize((256, 256), Image.ANTIALIAS)
        image.save('0000'+str(i)+'.jpg')
```
