```python
keyword = 'statue of liberty'

photos = flickr.walk(text=keyword,
                     tag_mode='all',
                     tags=keyword,
                     extras='url_c',
                     per_page=100,           # may be you can try different numbers..
                     sort='relevance')
```

```python
urls = []
for i, photo in enumerate(photos):
    print (i)
    url = photo.get('url_c')
    urls.append(url)
    
    # get 50 urls
    if i > 1000:
        break
```

```python
print (urls)

# Download image from the url and save it to '00001.jpg'
for i in range(0,900):
    if(urls[i] != None):
        urllib.request.urlretrieve(urls[i], '0000'+str(i)+'.jpg')
        # Resize the image and overwrite it
        image = Image.open('0000'+str(i)+'.jpg') 
        image = image.resize((256, 256), Image.ANTIALIAS)
        image.save('0000'+str(i)+'.jpg')
```
